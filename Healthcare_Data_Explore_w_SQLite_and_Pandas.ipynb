```python
import pandas as pd
import sqlite3
```

```python
# Step 1: create a connection to SQLite database
conn = sqlite3.connect('test_database.db')

```

```python
# Step 2: Create a Pandas data frame
df = pd.read_csv('U.S._Chronic_Disease_Indicators__CDI___2023_Release.csv')
```

```python
df.head()
# testing if the data imported correctly
```

```python
# save the dataframe to the SQLite database (step 2 in HW assignment, step 3 in example)
df.to_sql('chronic_disease_indicators', conn, if_exists='replace', index=False)
```

```python
# trying a query
query = "SELECT * FROM chronic_disease_indicators LIMIT 5;"
result = pd.read_sql_query(query, conn)
print(result)
```

```python
# another query test
query = "SELECT DISTINCT YearStart FROM chronic_disease_indicators;"
result = pd.read_sql_query(query, conn)
print(result)
```

```python
# Query 1
query = "SELECT DISTINCT LocationDESC FROM chronic_disease_indicators;"
result = pd.read_sql_query(query, conn)
print(result)

#
```

```python
# Query 2 : Count the number of records that meet a certain condition -
query = "SELECT COUNT(*) FROM chronic_disease_indicators WHERE Topic = 'Alcohol';"
result = pd.read_sql_query(query, conn)
print(result)
```

```python
# query that failed , another one for Query 2
query = "SELECT COUNT(*) FROM chronic_disease_indicators WHERE QUESTION = 'Mortality from heart failure';"
result = pd.read_sql_query(query, conn)
print(result)
```

```python
# query 3: group the data by a specific column and calculate a summary statistic for each group
query = "SELECT LocationDesc, COUNT(*) FROM chronic_disease_indicators GROUP BY LocationDesc;"
result = pd.read_sql_query(query, conn)
print(result)
```

```python
# query 4: sort the records based on a numerical or categorial field and return a limited set of results (like top 5)
query = "SELECT * FROM chronic_disease_indicators ORDER BY YearStart DESC LIMIT 5;"
result = pd.read_sql_query(query, conn)
print(result)
```
